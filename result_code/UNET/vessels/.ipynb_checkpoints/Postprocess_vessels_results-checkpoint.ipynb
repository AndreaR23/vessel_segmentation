## Try sharpen with medfilt on predicted images

```python
from scipy.signal import medfilt2d
from skimage import data, color
from skimage import measure
```

```python
imgs_RESNET18_dice, masks_RESNET18_dice, predicted_imgs_RESNET18_dice = predict_images(backbone='resnet18', 
                                                                                      val_frame_path=val_frame_path,
                                                                                      val_mask_path=val_mask_path,
                                                                                      img_height=img_height,
                                                                                      img_width=img_width,
                                                                                      model=modelRESNET18_dice,
                                                                                      preprocess=True)
# plot_images(imgs_RESNET18_dice, masks_RESNET18_dice, predicted_imgs_RESNET18_dice)
```

```python
def med_ext_med_filt(predicted_image, kernel_size=3):
    kernel = np.ones((4,4),np.uint8)

    img_squeezed = np.squeeze(predicted_image)   
    med_filter_img = medfilt2d(img_squeezed, kernel_size)
    
    return  medfilt2d(cv2.morphologyEx(med_filter_img, cv2.MORPH_CLOSE, kernel), 7)
```

```python
filtered_images = []

for im in predicted_imgs_RESNET18_dice:
    med_im = med_ext_med_filt(im, 3)
    med_im[med_im>0.01] = 1
    med_im[med_im<=0.01] = 0
#     filtered_images.append(cv2.GaussianBlur(med_im, (5,5),0))
    filtered_images.append(med_im)
```

```python
def draw_found_contours(origin_images, filtered_images, masks):
    for idx in range(len(origin_images)):
#         contours = measure.find_contours(filtered_images[idx], 0.8)

        plt.figure()
        fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(15,10))

        ax1.set_title('Image')
        ax1.imshow(origin_images[idx])
    
        ax2.set_title('Mask')
        ax2.imshow(np.squeeze(masks[idx]), cmap='gray')
        
        ax3.set_title('Predicted image')
        ax3.imshow(np.squeeze(filtered_images)[idx], cmap='gray')
```

```python
# draw_found_contours(imgs_RESNET18_dice, filtered_images, masks_RESNET18_dice)
draw_found_contours(imgs_RESNET18_dice, predicted_imgs_RESNET18_dice, masks_RESNET18_dice)
```
